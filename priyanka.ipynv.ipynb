#  Price vs Area (Scatter Plot) :-
This scatter plot shows how house prices change based on the area of the house. Each point represents a house, with area on the x-axis and price on the y-axis. Blue dots are used to indicate the houses, and a black edge color enhances the visibility of individual points.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

housing_data = pd.read_csv('Housing.csv')

plt.figure(figsize=(8, 6))
sns.scatterplot(x='area', y='price', data=housing_data, color='blue', edgecolor='black')
plt.title('Price vs Area')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.show()
```

# Price vs Bedrooms (Box Plot) :-
This box plot shows the distribution of house prices based on the number of bedrooms. Each box represents the range of prices for houses with a certain number of bedrooms, using different colors from the Set2 palette. The box plot highlights the median price and the spread (quartiles) of the data.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='bedrooms', y='price', data=housing_data, hue='furnishingstatus', palette='Set2')
plt.title('Price vs Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

# Price Distribution (Histogram with Density Plot):-
This histogram shows the overall distribution of house prices. The bars represent the number of houses within certain price ranges, and the green line indicates the density estimate (i.e., a smoothed curve showing the probability distribution). This graph helps understand the common price ranges for houses.

```python
housing_data = pd.read_csv('Housing.csv')

plt.figure(figsize=(8, 6))
sns.histplot(housing_data['price'], kde=True, color='green')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

#  Price vs Parking (Bar Plot) :-
This bar plot shows how the average house price changes depending on the number of parking spaces. The viridis color palette is used for better color contrast. The graph shows whether houses with more parking spots tend to have higher prices.

```python
housing_data = pd.read_csv('Housing.csv')
plt.figure(figsize=(8, 6))
sns.barplot(x='stories', y='price', hue='furnishingstatus', data=housing_data, palette='coolwarm')
plt.title('Price vs Stories')
plt.xlabel('Number of Stories')
plt.ylabel('Price')
plt.show()
```

#  Price vs Guestroom (Count Plot) :-
This count plot visualizes the number of houses that have a guestroom (yes) versus those that don't (no). The cubehelix palette is used to show the count of houses with different guestroom availability. This graph gives a quick overview of how many houses offer a guestroom feature.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='guestroom', data=housing_data, hue='furnishingstatus', palette='cubehelix')
plt.title('Count of Houses with Guestroom')
plt.xlabel('Guestroom Available')
plt.ylabel('Count')
plt.show()
```

# Air Conditioning Availability (Pie Chart) :-
This pie chart shows the proportion of houses that have air conditioning versus those that don't. The chart is divided into slices representing the percentage of houses in each category, providing a quick overview of air conditioning availability in the dataset.

```python
housing_data = pd.read_csv('Housing.csv')
airconditioning_counts = housing_data['airconditioning'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(airconditioning_counts, labels=airconditioning_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen'])
plt.title('Air Conditioning Availability')
plt.show()
```

```python
import pandas as pd

housing_data = pd.read_csv('Housing.csv')

average_price = housing_data['price'].mean()

max_price = housing_data['price'].max()
min_price = housing_data['price'].min()


house_max_area = housing_data.loc[housing_data['area'].idxmax()]

average_bedrooms = housing_data['bedrooms'].mean()
average_bathrooms = housing_data['bathrooms'].mean()

common_bedrooms = housing_data['bedrooms'].mode()[0]
common_bathrooms = housing_data['bathrooms'].mode()[0]

num_houses_with_ac = housing_data[housing_data['airconditioning'] == 'yes'].shape[0]

num_houses_with_guestroom = housing_data[housing_data['guestroom'] == 'yes'].shape[0]

num_houses_with_basement = housing_data[housing_data['basement'] == 'yes'].shape[0]

average_price_furnished = housing_data[housing_data['furnishingstatus'] == 'furnished']['price'].mean()
average_price_semi_furnished = housing_data[housing_data['furnishingstatus'] == 'semi-furnished']['price'].mean()
average_price_unfurnished = housing_data[housing_data['furnishingstatus'] == 'unfurnished']['price'].mean()

print("Important Data from the Housing Project:")
print(f"1. Average House Price: {average_price}")
print(f"2. Maximum House Price: {max_price}")
print(f"3. Minimum House Price: {min_price}")
print(f"4. House with Maximum Area:\n{house_max_area}")
print(f"5. Average Number of Bedrooms: {average_bedrooms}")
print(f"6. Average Number of Bathrooms: {average_bathrooms}")
print(f"7. Most Common Number of Bedrooms: {common_bedrooms}")
print(f"8. Most Common Number of Bathrooms: {common_bathrooms}")
print(f"9. Number of Houses with Air Conditioning: {num_houses_with_ac}")
print(f"10. Number of Houses with Guestroom: {num_houses_with_guestroom}")
print(f"11. Number of Houses with Basement: {num_houses_with_basement}")
print(f"12. Average Price for Furnished Houses: {average_price_furnished}")
print(f"13. Average Price for Semi-Furnished Houses: {average_price_semi_furnished}")
print(f"14. Average Price for Unfurnished Houses: {average_price_unfurnished}")
```
